```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
demodata = pd.read_csv('US_county_condensed.csv')
demodata.head()
```

```python
demodata["Case density (Per 100000)"] = (demodata["Confirmed"]/demodata["Total Population"])*100000
demodata["Death density (Per 100000)"] = (demodata["Deaths"]/demodata["Total Population"])*100000
demodata["Death Rate (%)"] = (demodata["Deaths"]/demodata["Confirmed"])*100
```

```python
demodata.head(20)
```

```python
f = plt.figure(figsize=(20, 20))
sb.heatmap(demodata.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 8}, cmap = "RdBu")
```
