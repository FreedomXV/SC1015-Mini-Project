```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
demodata = pd.read_csv('county_megaset.csv')
demodata.head()
```

```python
print("Data type : ", type(demodata))
print("Data dims : ", demodata.shape)
```

```python
poverty = pd.DataFrame(demodata[["Percentage of people whose income in the past 12 months is below the poverty level","Confirmed","Deaths"]])
```

```python
poverty.describe()
```

```python
poverty.head(20)
```

```python
# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(20, 20))
sb.heatmap(poverty.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
sb.pairplot(data = poverty)
```

```python
US_county_condensed = pd.DataFrame(demodata[["Geographic Identifier - FIPS Code","Name","State","Area of Land (square meters)","Area of Water (square meters)",
                                             "Average Household Size","Total Population","Population Density (people per square kilometer)",
                                             "Percent of Population with No Health Insurance Coverage","Households: Receiving Food Stamps/SNAP (%)",
                                             "Households: Income Below Poverty Level (%)","Lat","Long_",
                                             "Percentage of people whose income in the past 12 months is below the poverty level","Confirmed","Deaths"]])
```

```python
US_county_condensed.head()
```

```python
f = plt.figure(figsize=(20, 20))
sb.heatmap(US_county_condensed.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 8}, cmap = "RdBu")
```
