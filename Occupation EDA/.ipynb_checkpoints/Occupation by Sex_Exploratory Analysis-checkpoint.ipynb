```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
occup_data = pd.read_csv('Occupation_condensed.csv')
occup_data.head()
```

```python
occup_data.describe()
```

Now we will examine each of the data sets one-by-one to find data points that influences the covid-19 cases confirmed and also deaths in the counties. We will first plot the box and whiskers plot, histogram and violin plot of the Confirmed cases and death cases.

```python
f, axes = plt.subplots(2, 3, figsize=(24, 12))

sb.boxplot(data = occup_data['Confirmed'], orient = "h", ax = axes[0,0])
sb.histplot(data = occup_data['Confirmed'], ax = axes[0,1])
sb.violinplot(data = occup_data['Confirmed'], orient = "h", ax = axes[0,2])

sb.boxplot(data = occup_data['Deaths'], orient = "h", ax = axes[1,0])
sb.histplot(data = occup_data['Deaths'], ax = axes[1,1])
sb.violinplot(data = occup_data['Deaths'], orient = "h", ax = axes[1,2])
```

From the above results we can see that it is very weird and there is no results that we can infer from the results above mostly due to the amount of outliers that is present in the dataset. Therefore, we have to remove the outliers.

```python
#Removing Outliers
Q1c = occup_data['Confirmed'].quantile(0.25)
Q3c = occup_data['Confirmed'].quantile(0.75)
IQRc = Q3c - Q1c
occup_data_wooutliers = pd.DataFrame(occup_data[((occup_data['Confirmed'] > (Q1c - 1.5 * IQRc)) & (occup_data['Confirmed'] < (Q3c + 1.5 * IQRc)))])
```

```python
#Removing Outliers
Q1d = occup_data['Deaths'].quantile(0.25)
Q3d = occup_data['Deaths'].quantile(0.75)
IQRd = Q3d - Q1d
occup_data_wooutliers = pd.DataFrame(occup_data_wooutliers[((occup_data_wooutliers['Deaths'] > (Q1d - 1.5 * IQRd)) & (occup_data_wooutliers['Deaths'] < (Q3d + 1.5 * IQRd)))])
```

We will remove the outliers for our datapoints too.

```python
#Removing Outliers
Q1d = occup_data['Total - Civilian employed population 16 years and over'].quantile(0.25)
Q3d = occup_data['Total - Civilian employed population 16 years and over'].quantile(0.75)
IQRd = Q3d - Q1d
occup_data_wooutliers = pd.DataFrame(occup_data_wooutliers[((occup_data_wooutliers['Total - Civilian employed population 16 years and over'] > (Q1d - 1.5 * IQRd)) & (occup_data_wooutliers['Total - Civilian employed population 16 years and over'] < (Q3d + 1.5 * IQRd)))])
```

Now that we have removed the outliers from the dataset, we can now plot the box and whiskers plot, histogram and violin plot of the dataset without outliers. 

```python
f, axes = plt.subplots(2, 3, figsize=(24, 12))

sb.boxplot(data = occup_data_wooutliers['Confirmed'], orient = "h", ax = axes[0,0])
sb.histplot(data = occup_data_wooutliers['Confirmed'], ax = axes[0,1])
sb.violinplot(data = occup_data_wooutliers['Confirmed'], orient = "h", ax = axes[0,2])

sb.boxplot(data = occup_data_wooutliers['Deaths'], orient = "h", ax = axes[1,0])
sb.histplot(data = occup_data_wooutliers['Deaths'], ax = axes[1,1])
sb.violinplot(data = occup_data_wooutliers['Deaths'], orient = "h", ax = axes[1,2])
```

Now this looks much better, although the shapes are still weird, but it is normal because the some of the counties are relatively very small compared to the other counties.

We will now plot a joint plot of the data points versus the confirmed cases and deaths cases, more specifically Confirmed cases and Death cases versus Total - Civilian employed population 16 years and over. However, we will start plotting using the datasets with outliers then the datasets without the outliers to better show the difference in correlation later on.


```python
sb.jointplot(data = occup_data, x = "Confirmed", y = "Total - Civilian employed population 16 years and over", height = 8)
sb.jointplot(data = occup_data, x = "Deaths", y = "Total - Civilian employed population 16 years and over", height = 8)
```

```python
sb.jointplot(data = occup_data_wooutliers, x = "Confirmed", y = "Total - Civilian employed population 16 years and over", height = 8)
sb.jointplot(data = occup_data_wooutliers, x = "Deaths", y = "Total - Civilian employed population 16 years and over", height = 8)
```

As we can see, the difference with and without outliers is really huge, at least we can observe some pattern in the joint plot without outliers. We will now move on to the most important part, which is to calculate the correlation between the datapoints.

```python
occup_data1 = occup_data[['Deaths', 'Confirmed', 'Total - Civilian employed population 16 years and over']]
occup_data1.corr()
```

```python
occup_data_wooutliers1 = occup_data_wooutliers[['Deaths', 'Confirmed', 'Total - Civilian employed population 16 years and over']]
occup_data_wooutliers1.corr()
```

```python
#Removing Outliers
Q1d = occup_data['Total Healthcare-related workers'].quantile(0.25)
Q3d = occup_data['Total Healthcare-related workers'].quantile(0.75)
IQRd = Q3d - Q1d
occup_data_wooutliers = pd.DataFrame(occup_data_wooutliers[((occup_data_wooutliers['Total Healthcare-related workers'] < (Q3d + 1.5 * IQRd)))])
```

```python
occup_data2 = occup_data[['Deaths', 'Confirmed', 'Total Healthcare-related workers']]
occup_data2.corr()
```
