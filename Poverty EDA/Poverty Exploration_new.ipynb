```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
demodata = pd.read_csv('US_county_condensed.csv')
demodata.head()
```

```python
demodata["Case density (Per 100000)"] = (demodata["Confirmed"]/demodata["Total Population"])*100000
demodata["Death density (Per 100000)"] = (demodata["Deaths"]/demodata["Total Population"])*100000
demodata["Death Rate (%)"] = (demodata["Deaths"]/demodata["Confirmed"])*100
```

```python
demodata.head(20)
```

```python
f = plt.figure(figsize=(20, 20))
sb.heatmap(demodata.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 8}, cmap = "RdBu")
```

### Basic Exploration

```python
print('Summary Statistics of [Households: Income Below Poverty Level (%)], [Case density (Per 100000)],[Death density (Per 100000)], [Death Rate (%)]')
Povdata = demodata[['Households: Income Below Poverty Level (%)','Case density (Per 100000)','Death density (Per 100000)','Death Rate (%)']]
Povdata.describe()
```

```python
print('Uni-variate Subplots')
# Set up matplotlib figure with three subplots
f, axes = plt.subplots(3, 3, figsize=(24, 12))

sb.boxplot(data = Povdata['Households: Income Below Poverty Level (%)'], orient = "h", ax = axes[0,0])
sb.histplot(data = Povdata['Households: Income Below Poverty Level (%)'], ax = axes[0,1])
sb.violinplot(data = Povdata['Households: Income Below Poverty Level (%)'], orient = "h", ax = axes[0,2])

sb.boxplot(data = Povdata['Case density (Per 100000)'], orient = "h", ax = axes[1,0])
sb.histplot(data = Povdata['Case density (Per 100000)'], ax = axes[1,1])
sb.violinplot(data = Povdata['Case density (Per 100000)'], orient = "h", ax = axes[1,2])

sb.boxplot(data = Povdata['Death density (Per 100000)'], orient = "h", ax = axes[2,0])
sb.histplot(data = Povdata['Death density (Per 100000)'], ax = axes[2,1])
sb.violinplot(data = Povdata['Death density (Per 100000)'], orient = "h", ax = axes[2,2])
```

```python
# Jointplot of Poverty against Case incidence
sb.jointplot(data = Povdata, x = 'Households: Income Below Poverty Level (%)', y = 'Case density (Per 100000)', height = 12)
print('Jointplot of Poverty against Case incidence')
```

```python
# Jointplot of Poverty against Death incidence
sb.jointplot(data = Povdata, x = 'Households: Income Below Poverty Level (%)', y = 'Death density (Per 100000)', height = 12)
print('Jointplot of Poverty against Case incidence')
```

```python
# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(20, 20))
sb.heatmap(Povdata.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
# Import essential models and functions from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Creating new datasets for predictor and response variables
Case = pd.DataFrame(Povdata['Case density (Per 100000)'])        # Response
Poverty = pd.DataFrame(Povdata['Households: Income Below Poverty Level (%)'])         # Predictor

# Splitting the Dataset into Train and Test in an 80:20 ratio randomly
X_train, X_test, y_train, y_test = train_test_split(Poverty, Case, test_size = 0.20)

# Linear Regression using Train Data
linreg = LinearRegression()         # create the linear regression object
linreg.fit(X_train, y_train)        # train the linear regression model

# Coefficients of the Linear Regression line
print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()

# Formula for the Regression line
regline_x = X_train
regline_y = linreg.intercept_ + linreg.coef_ * X_train

# Plot the Linear Regression line
f = plt.figure(figsize=(16, 8))
plt.scatter(X_train, y_train)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()
```

```python
# Predict SalePRice values corresponding to GrLivArea
y_train_pred = linreg.predict(X_train)
y_test_pred = linreg.predict(X_test)
```

```python
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_train, y_train))
print()
```

```python
# Check the Prediction Accuracy (on Test Data)
print("Prediction Accuracy \tTest Dataset")
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print()
```

#### Attempt without the outliers

We will now attempt the regression of poverty rate against case density without the outliers for both variables

```python
Povdata_clean = demodata[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']]
```

```python
#Checking the vital statistics of the dataset before removal of outliers
print("Data type : ", type(Povdata_clean))
print("Data dims : ", Povdata_clean.shape)
```

```python
print('Summary Statistics of Households: Income Below Poverty Level (%) and Case density (Per 100000) before cleaning')
Povdata_clean[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']].describe()
```

```python
#removing outliers
Q1 = Povdata_clean[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']].quantile(0.25)
Q3 = Povdata_clean[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']].quantile(0.75)
IQR = Q3-Q1
upper = np.where(Povdata_clean[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']] > 1.5*IQR[[0,1]]+Q3[[0,1]])
lower = np.where(Povdata_clean[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']] < Q1[[0,1]]-1.5*IQR[[0,1]])
Povdata_clean.drop(upper[0], inplace = True)
Povdata_clean.drop(lower[0], inplace = True)
```

```python
print('New data dimensions')
print("Data type : ", type(Povdata_clean))
print("Data dims : ", Povdata_clean.shape)
```

```python
print('Summary Statistics of Households: Income Below Poverty Level (%) and Case density (Per 100000) after cleaning')
Povdata_clean[['Households: Income Below Poverty Level (%)','Case density (Per 100000)']].describe()
```

```python
# Import essential models and functions from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Creating new datasets for predictor and response variables
Case2 = pd.DataFrame(Povdata_clean['Case density (Per 100000)'])        # Response
Poverty2 = pd.DataFrame(Povdata_clean['Households: Income Below Poverty Level (%)'])         # Predictor

# Splitting the Dataset into Train and Test in an 80:20 ratio randomly
X_train2, X_test2, y_train2, y_test2 = train_test_split(Poverty2, Case2, test_size = 0.20)

# Linear Regression using Train Data
linreg2 = LinearRegression()         # create the linear regression object
linreg2.fit(X_train2, y_train2)        # train the linear regression model

# Coefficients of the Linear Regression line
print('Intercept of Regression \t: b = ', linreg2.intercept_)
print('Coefficients of Regression \t: a = ', linreg2.coef_)
print()

# Formula for the Regression line
regline_x2 = X_train2
regline_y2 = linreg2.intercept_ + linreg2.coef_ * X_train2

# Plot the Linear Regression line
f = plt.figure(figsize=(16, 8))
plt.scatter(X_train2, y_train2)
plt.plot(regline_x2, regline_y2, 'r-', linewidth = 3)
plt.show()
```

```python
# Predict SalePRice values corresponding to GrLivArea
y_train_pred2 = linreg2.predict(X_train2)
y_test_pred2 = linreg2.predict(X_test2)
```

```python
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg2.score(X_train2, y_train2))
print()
```

```python
# Check the Prediction Accuracy (on Test Data)
print("Prediction Accuracy \tTest Dataset")
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test2, y_test_pred2))
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mean_squared_error(y_test2, y_test_pred2)))
print()
```

With outliers removed, there is some improvement in the model, though insignificant.
